```python
#importing the dependencies 

import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('credit_data.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
#determine the distribution of the 2 classes
dataset['Class'].value_counts()
```

```python
#seperating the data legit and fraud

legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
#Building the sample dataset containing similar distribution of legit and fraudulent transactions

legit_sample = legit.sample(n=492)
print(legit_sample.shape)
```

```python
#Concatenate the Two Dataframes

new = pd.concat([legit_sample, fraud],axis = 0)
```
