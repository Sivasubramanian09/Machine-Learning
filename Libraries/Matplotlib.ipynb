```python
#To Plot Parabola

x= np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel("X Value"), plt.ylabel("Y Value")
plt.title("Parabola Graph")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English','French','Spanish','German','Tamil']
people = [100,40,60,25,39]
ax.bar(language,people)
plt.xlabel("Languages"),plt.ylabel("No.of People")
```

```python
#pie chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','German']
people = [10,30,60]
ax.pie(people, labels = languages ,autopct = '%1.1f%%')
plt.show()
```

```python
#scatter plot

x = np.linspace(10,20,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'b')
plt.show()
```

```python
#3D SCATTER PLOT

fig3 = plt.figure()
ax = plt.axes(projection='3d')
z=20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,c=z,cmap = 'Blues')
plt.show
```
