```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#No te : Seaborn has some Built in Dataset

tips = sns.load_dataset('tips')
tips.head()
```

```python
#visualize the data
sns.set_theme()
sns.relplot(data = tips, x='total_bill', y ='tip',col='time',hue = 'smoker',style = 'smoker',size = 'size')
plt.show()
```

```python
# load the Iris dataset

iris = sns.load_dataset('iris')

iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_width',hue='species',data=iris)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.relplot(data = titanic,x='age',y='fare',hue = 'sex')
```

```python
sns.countplot(x='age',data = titanic)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data = titanic)
```

```python
sns.displot(titanic['fare'])
```

```python
#Corelation Map

correlation = titanic.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar= True, square=True,fmt='.1f', annot = True, annot_kws={'size':8},cmap='Blues')
```
