```python
#Vectors Operations in python
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
#Plotting a Vector

plt.quiver(0,0,4,5, scale_units = 'xy' , angles = 'xy', scale =1,color='b')
plt.quiver(0,0,-3,-6, scale_units = 'xy' , angles = 'xy', scale =1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Addition of Vectors

vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy' , angles = 'xy', scale =1,color='b')
plt.quiver(0,0,3,-2, scale_units = 'xy' , angles = 'xy', scale =1,color='y')
plt.quiver(0,0,5,1, scale_units = 'xy' , angles = 'xy', scale =1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Subtraction of Vectors

vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 - vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy' , angles = 'xy', scale =1,color='b')
plt.quiver(0,0,3,-2, scale_units = 'xy' , angles = 'xy', scale =1,color='y')
plt.quiver(0,0,-1,5, scale_units = 'xy' , angles = 'xy', scale =1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Multiplication of Vector
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy' , angles = 'xy', scale =1,color='b')
plt.quiver(0,0,4,6, scale_units = 'xy' , angles = 'xy', scale =1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Dot Products of 2 Vectors
```

```python
a = np.array([2,3])

b = np.array([4,4])

a_dot = np.dot(a,b)
```

```python
print(a_dot)
```

```python
a = np.array([10,20,30])
b = np.array([3,4,5])
dot = np.dot(a,b)
```

```python
print(dot)
```

```python
#Cross Product
```

```python
a = np.array([2,3])

b = np.array([4,4])

cross = np.cross(a,b)
```

```python
print(cross)
```

```python
c = np.array([5,10,20])

d = np.array([18,32,50])

cross_product = np.cross(c,d)
```

```python
print(cross_product)
```

```python
#Projection of Vector

a = np.array([2,5])

v = np.array([8,-6])

#magnitude of 'v' vector

magnitude_of_v = np.sqrt(np.sum(v**2))

proj = ((np.dot(a,v))/((magnitude_of_v)**2))*v
```

```python
print(proj)
```
